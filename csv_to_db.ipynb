```python
import pandas as pd
from sqlalchemy import create_engine
print("imported important stuff!")
```

```python
engine = create_engine(
    'mysql+mysqlconnector://localhost/<<database_name>>',
    connect_args={
        'user': '<<username>>',
        'password': '<<password>>'
    }
)
print("connected important sql stuff!")
```

```python
df = pd.read_csv('ipl_2026_final_bbb.csv')
df.to_sql('deliveries', con=engine, if_exists='replace', index=False)
print('Done')
```

```python
df_Ba = pd.read_csv('batter_averages_v2.csv')
df_Ba.to_sql('batter_averages', con=engine, if_exists='replace', index=False)
print('Done')
```

```python
df_ven = pd.read_csv('venue_details_clean.csv')
df_ven.to_sql('venue_details', con=engine, if_exists='replace', index=False)
print('Done')
```
